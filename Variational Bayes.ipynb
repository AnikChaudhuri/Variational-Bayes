```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(5,2,200)
xbar = np.mean(x)
x_square = np.sum(x*x)
lambda0=1; mu0=2; a0=.3; b0 = 2; lambdan=3
mu_n = (lambda0*mu0 + len(x)*xbar)/(lambda0 + len(x))
mean = np.zeros(1000)
var = np.zeros(1000)
an = a0+(len(x)+1)/2
for i in range(1000):
    
    bn = b0 +0.5*((lambda0+len(x))*(1/lambdan+mu_n**2)-2*(lambda0*mu0+np.sum(x))*mu_n+x_square+lambda0*mu0**2)
    lambdan = (lambda0+len(x))*(an/bn)
    mean[i] = np.random.normal(mu_n,1/lambdan)
    var[i] = np.random.gamma(an,1/bn)
    
```

```python
plt.plot(mean)
plt.figure()
plt.plot(np.sqrt(1/var))
```
